```python
import pandas as pd
# 결측치 여부 확인(True: 결측치가 있다, False:값이 있다)
#DataFrame.isnull()

#데이터프레임생성
data = {
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,None,30],
    'Score':[90,85,None]
}

df = pd.DataFrame(data)
df.isnull()

#컬럼별 결측치 개수 확인
print("결측치 확인")
print(df.isnull().sum())

#결측치가 있는 행 삭제
df_cleaned = df.dropna()
print("결측치 삭제 후 데이터")
print(df_cleaned)
```

```python
#데이터프레임생성
data2 = {
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,None,30],
    'Score':[90,85,None]
}

df2 = pd.DataFrame(data2)
df2

#Age열의 평균값으로 대체
mean_age = df2['Age'].mean()
df2['Age'] = df2['Age'].fillna(mean_age)
df2
```

```python
#특정값으로 결측치 대체
df2['Score'] = df2['Score'].fillna(0)
df2
```
