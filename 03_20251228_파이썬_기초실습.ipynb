```python
#데이터베이스 파일 불러오기
import sqlite3
#SQLite 데이터베이스 연결 및 데이터 읽기
local_file = '/content/drive/MyDrive/Chinook_Sqlite.sqlite'
conn = sqlite3.connect(local_file) #데이터베이스 연결

#sqlite 데이터베이스의 테이블 리스트 확인하기
tables_df = pd.read_sql_query("""SELECT name AS table_name FROM sqlite_master WHERE type ='table' ORDER BY name""",conn) #쿼리 결과를 데이터프레임으로 변환
tables_df
#customer 테이블 데이터 불러오기
query = 'SELECT * FROM Customer'
df_customers = pd.read_sql_query(query,conn)
conn.close() #데이터베이스 연결 종료

df_customers.head()
```

```python
#HTML 파일 불러오기
import requests

# pd.read_html(url)이 내부적으로 urllib로 페이지를 가져오는데, 위키피디아가 Colab에서 오는 “기본 User-Agent(봇/스크립트)” 요청을 차단해서
# HTTP Error 403: Forbidden 이 발생
# 해결은 요청 헤더(User-Agent)를 붙여서 HTML을 먼저 받아온 뒤, 그 HTML 문자열을 pd.read_html()에 넣는 방식으로 변경

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
}

#HTML 테이블 데이터 읽기 : <table></table>태그를 이용해 데이터를 테이블 형식으로 표현
response = requests.get(url, headers=headers)
response.raise_for_status() #오류면 여기서 바로 알려줌

tables = pd.read_html(response.text) #html 페이지의 모든 테이블 읽기

# 몇개의 테이블이 있는지 확인
print("테이블 갯수:", len(tables))

# 모든 테이블의 컬럼 이름만 한 번에 보기
for i,table in enumerate(tables):
  print(f"\n====테이블{i}====")
  print(table.columns)

gdp_table = tables[2]

gdp_table.head()
```
