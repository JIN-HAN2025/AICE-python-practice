# 동작설명
1. 다음 열들을 삭제
*   PassengerId
*   Name
*  Ticket
2. SibSp열과	Parch열의 값을 더해 Family라는 새로운 열 생성
3.Age열의 값이 결측치인 경우 Age열의 평균으로 대체
4.Cabin열의 값이 결측치인 경우 X라는 문자로 대체
5.Embarked 열의 값이 결측치인 경우 해당 행을 삭제
6.Cabin열 열에서 첫 글자만 Cabin2라는 열로 새로 입력


```python
df_titanic.head()
```

```python


df_titanic = pd.read_csv('/content/sample_data/titanic.csv') # Add this line to re-initialize df_titanic

# 1) 열 삭제
df_titanic = df_titanic.drop(columns=["PassengerId", "Name", "Ticket"])

# 2) SibSp + Parch => Family 열 생성
df_titanic["Family"] = df_titanic["SibSp"] + df_titanic["Parch"]

# 3) Age 결측치 => Age 평균으로 대체
age_mean = df_titanic["Age"].mean()
df_titanic["Age"] = df_titanic["Age"].fillna(age_mean)

# 4) Cabin 결측치 => 'X'로 대체
df_titanic["Cabin"] = df_titanic["Cabin"].fillna("X")

# 5) Embarked 결측치 행 삭제
df_titanic = df_titanic.dropna(subset=["Embarked"])

# 6) Cabin 첫 글자만 추출하여 Cabin2 열 생성
#    (Cabin이 'X'인 경우 Cabin2도 'X'가 됨)
df_titanic["Cabin2"] = df_titanic["Cabin"].astype(str).str[0]
df_titanic.head()
```
