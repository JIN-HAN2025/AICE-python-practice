```python
#딕셔너리{..}하나의 층에 방이 여러개 있는것
#딕셔너리 = 한 행(열이름, 값)
person = {'name':'홍길동',
          'age':43}
print(person)
print(person['name'])
```

```python
#한 사람의 정보를 키보드로 입력받는다.
#이름, 성별
#한 사람의 정보를 딕셔너리 형태로 만들어본다

name1 = input('이름?')
gender1 = input('성별?')
person1 = {'이름':name1,'성별':gender1}
print(person1)
```

```python
name2 = input('이름?')
gender2 = input('성별?')
person2 = {'이름':name2,'성별':gender2}
print(person2)
```

```python
# 딕셔저기 = 한행
# 리스트 = 여러행
persons = [person1,person2]
print(persons)
```
